```python
import pandas as pd
```

```python
df = pd.read_csv('../data/admissions_ibd.csv')

def parse_race(race):
    if 'WHITE' in race:
        return 'WHITE'
    elif 'BLACK' in race:
        return 'BLACK'
    elif 'HISPANIC' in race or 'LATINO' in race:
        return 'HISPANIC/LATINO'
    elif 'ASIAN' in race:
        return 'ASIAN'
    else:
        return 'OTHER'

df['race'] = df['race'].apply(parse_race)
```

```python
df.head()
```

```python
df['subject_id'].unique().shape
```

```python
grouped = df[['subject_id', 'admission_location', 'discharge_location', 'insurance', 'language', 'marital_status', 'race']].groupby('subject_id')
```

```python
stats = grouped.first()

stats
```

```python
for col in stats.columns:
    missing_value_nums = stats[stats[col].isna()].shape[0]
    print('missing value: {}'.format(missing_value_nums))
    summary = stats[[col]].value_counts()
    print(summary)
    print('\n')
```

```python
df2 = pd.read_csv('../data/patients_ibd.csv')
df2.head()
```

```python
for col in df2.columns[1:]:
    missing_value_nums = df2[df2[col].isna()].shape[0]
    print('missing value: {}'.format(missing_value_nums))
    summary = df2[[col]].value_counts()
    print(summary)
    print('\n')
```

# LOS

```python
df3['los'] = pd.to_datetime(df3['dischtime']) - pd.to_datetime(df3['admittime'])
```

```python
stats_los = df3[['subject_id', 'los']].groupby('subject_id').agg({
    'los': ['mean', 'min', 'max', 'std', 'count']
})

stats_los.columns = ['mean', 'min', 'max', 'std', 'count']
stats_los = stats_los.reset_index()
```

```python
stats_los['count'].mean()
```

```python
stats_los['mean'].mean()
```
