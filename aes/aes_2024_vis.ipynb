```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from utils import load_data, preprocess_icu_ibd
```

```python
icu_ibd = preprocess_icu_ibd(load_data('icu_ibd'))
```

```python
icu_ibd.columns
```

```python
for col in ['gender', 'race', 'language',
       'marital_status', 'insurance', 'ibd_type']:
    
    print(icu_ibd[col].value_counts())
```

```python
icu_ibd.columns
```

```python
# 
plt.figure(figsize=(7, 5), dpi=300)
sns.histplot(data=icu_ibd, x="age", kde=True)
plt.title('Age Distribution')
plt.show()
```

```python

plt.figure(figsize=(7, 5), dpi=300)
sns.histplot(data=icu_ibd, x="heart_rate", kde=True, color='red')
plt.title('Heart Rate Distribution')
plt.show()
```
